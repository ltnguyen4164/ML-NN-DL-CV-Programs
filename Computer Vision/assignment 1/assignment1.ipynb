# Image Transformations

```python
import numpy as np
# skimage used to open files
import skimage.io as io
import skimage.color as color
from skimage.util import img_as_ubyte
# matplotlib used to display images
import matplotlib.pyplot as plt
# assignment 1 programs
import img_transforms as tfs
import create_img_pyramid as cimg

# change filename to any file you want to test
filename = 'test.png'
img = io.imread(filename)

if img.ndim == 3 and img.shape[2] == 4:
  img = color.rgba2rgb(img)
```

# Random Cropping

```python
# change size to any value
size = 100
img_cropped = tfs.random_crop(img, size)
img_cropped = img_as_ubyte(img_cropped)
plt.imshow(img_cropped)
```

# Patch Extraction

```python
# change num_patches to any value
num_patches = 4
img_patches = tfs.extract_patch(img, num_patches)
img_patches = img_as_ubyte(img_patches)
# plot patches
fig, axes = plt.subplots(num_patches, num_patches, figsize=(5, 5))
for i in range(num_patches):
    for j in range(num_patches):
        axes[i, j].imshow(img_patches[i, j])
        axes[i, j].axis("off")

plt.tight_layout()
plt.show()
```

# Resizing

```python
# change scale_factor to any value > 0
scale_factor = 4
img_resized = tfs.resize_img(img, scale_factor)
img_resized = img_as_ubyte(img_resized)
plt.imshow(img_resized)
```

# Color Jitter

```python
# change hue, sat, and value to any value following the ranges below:
# hue must be in range (0, 360]
# sat and value must be in range (0, 1]
hue, sat, value = 0, 0.5, 0.5
img_jittered = tfs.color_jitter(img, hue, sat, value)
img_jittered = img_as_ubyte(img_jittered)
plt.imshow(img_jittered)
```
